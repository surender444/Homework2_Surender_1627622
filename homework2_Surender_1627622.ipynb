
1. Use for loop to generate 3 random integers between 100 and 999 which is divisible by 5. 



```python
# This is the answer for question 1
import random

def generate_random_numbers(min_value, max_value, divisible_by):
 

  numbers = []
  for i in range(3):
    number = random.randint(min_value, max_value)
    while number % divisible_by != 0:
      number = random.randint(min_value, max_value)
    numbers.append(number)

  return numbers


if __name__ == "__main__":
  # This is to Generate 3 random numbers 
  numbers = generate_random_numbers(100, 999, 5)

  # This is here to print the text which come right before the numbers.
  print("The Generated Randon numbers are:", numbers)
```

2.The Logistic distribution is used in Extreme Value problems where it can act as a mixture of Gumbel distributions. Draw samples from a logistic distribution and plot it. loc=10, scale=1 and size =1000. 

```python
# This is the answer for question 2
import numpy as np
import matplotlib.pyplot as plt

# This here is for the Mean, Standard deviation, and Number of samples to generate
samples = np.random.logistic(loc=10, scale=1, size=1000)


plt.hist(samples, bins=30, density=True, alpha=0.7)
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.title("Samples from Logistic Distribution")
plt.grid(True)

x = np.linspace(np.min(samples), np.max(samples), 100)
pdf = np.exp(-(x - 10) / 1) / (1 + np.exp(-(x - 10) / 1))**2
plt.plot(x, pdf, 'r-', label="PDF")

plt.legend()
plt.show()
```

6. Import string and random module. Generate a Password with length 10 which contains 2 upper case letters, 1 digit, and 1 special symbol. 

```python
# This is the answer for question 6
import string
import random

# This is to define the set of characters for each category
upper_chars = string.ascii_uppercase
digit_chars = string.digits
special_chars = string.punctuation

# This is to generate the password
password = random.choices(upper_chars, k=2) + random.choices(digit_chars) + random.choices(special_chars) + random.choices(string.ascii_letters + string.digits + string.punctuation, k=6)
random.shuffle(password)
password = ''.join(password)

# This is to print the password
print("Generated password:", password)
```

7. First generate evenly spaced values within 0 to 9. Give it new shape 3x3. Then randomly permute it.

```python
# This is the answer for question 7
import numpy as np

# This is to generate evenly spaced values from 0 to 9
values = np.linspace(0, 9, 9, dtype=int)
print("Given values for Program:\n", values)

#  This is to reshape the values into a 3x3 matrix
matrix = np.reshape(values, (3, 3))
print(" Created matrix:\n", matrix)

# This is to randomly permute the elements
permuted_matrix = np.random.permutation(matrix)
print("Permuted matrix:\n", permuted_matrix)
```

10. Write an interactive Python program to play a “guess the  number” game. The computer should think of a random integer between  1 and 20 and the user (player) has to try to guess the number within six  attempts. The program should let the player know if the guess is too high or  too low. 


```python
# This is the answer for question 10
import random

def guess_the_number():
    # This is to generate a random number between 1 and 20
    secret_number = random.randint(1, 20)

    print("Welcome to the Guess the Number game!")
    print("I'm thinking of a number between 1 and 20.")

    attempts = 0
    while attempts < 6:
        print("\nAttempt:", attempts + 1)

        # This is to get the user's guess
        guess = int(input("Take a guess: "))

        if guess < secret_number:
            print("Too low!")
        elif guess > secret_number:
            print("Too high!")
        else:
            print("Congratulations! You guessed the number correctly!")
            return

        attempts += 1

    print("\nGame Over!")
    print("The secret number was:", secret_number)


guess_the_number()
```
